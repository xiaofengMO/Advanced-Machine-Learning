# A5

```python
def get_store_dict(str_pattern):
    files = glob.glob(str_pattern)
    highest_file = ""
    highest_score = 0
    for f in files:
        store_dict = np.load(f).item()
        if max(store_dict['test_preformance_list']) > highest_score:
            highest_score = max(store_dict['test_preformance_list'])
            highest_file = f
    print(highest_file)
    store_dict = np.load(highest_file).item()
    return store_dict
```

```python
store_dict = get_store_dict('../models/A_5/A_5_*_30_*.npy')
plt.figure()
plt.subplot(311)
plt.ylabel("eposide length")
plt.plot(store_dict["test_preformance_list"])
plt.subplot(312)
plt.ylabel("discounted return")
plt.plot(store_dict["test_dis_return"])

plt.subplot(313)
plt.plot(store_dict["bellman_loss_list"])
plt.ylabel("bellman loss")

plt.xlabel("evaluation per 20 steps during training, A 5 units = 30")

savefig("./A_5_image/A_5_30_units.png")
b = "\\caption{" + "A_5_30_units".replace("_", " ") + "}\n"
b += "\\includegraphics[scale=.8]{" + "A_5_30_units.png" + "}\n\\centering"
print(a + b + c)
```
