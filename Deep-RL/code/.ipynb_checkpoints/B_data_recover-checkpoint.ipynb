```python
import random
import numpy as np
import matplotlib.pyplot as plt
import os,sys,time
from pylab import savefig

# disable tensorflow debugging information
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# choose which game to run by system arguments
# game_name = "Pong"
# game_name = 'MsPacgirl'
# game_name = 'Boxing'
a = """
\\begin{figure}[h]
"""

c = """
\\end{figure}
"""
for game_name in ["Pong", "MsPacgirl", "Boxing"]:
    # print the game_name
    print(game_name)

    # get the file name for storing purposes
    file_name = "B_4_3"
    # setup a path for this script by the file name and game name
    path = "../models/{}/{}".format(file_name,game_name)

    step_loss = list(np.load(path + "/" + game_name + "-step_loss.npy"))
    test_perform = list(np.load(path + "/" + game_name + "-test_perform.npy"))
    test_rtn = list(np.load(path + "/" + game_name + "-test_rtn.npy"))

    plt.figure()
    plt.subplot(311)
    plt.ylabel("loss")
    plt.plot(step_loss[::50000])
    plt.subplot(312)
    plt.plot(step_loss[::50000][2:])
    
#     plt.plot(step_loss[:][50000 * 2:])

    plt.ylabel("loss zoom in")
    plt.xlabel("evaluation every 50k, {}".format(game_name))
    savefig("./B_image/B_{}_loss.png".format(game_name))
    b = "\\caption{" + "B_{}_loss".format(game_name).replace("_", " ") + "}\n"
    b += "\\includegraphics[scale=.8]{" + "B_{}_loss.png".format(game_name) + "}\n\\centering"
    print(a + b + c)

    plt.figure()
    plt.subplot(311)
    plt.ylabel("score")
    plt.plot(test_perform)
    plt.subplot(312)
    plt.ylabel("discounted rewaard")
    plt.plot(test_rtn)
    plt.xlabel("evaluation during training, {}".format(game_name))
    savefig("./B_image/B_{}_test_performance.png".format(game_name))
    b = "\\caption{" + "B_{}_test_performance".format(game_name).replace("_", " ") + "}\n"
    b += "\\includegraphics[scale=.8]{" + "B_{}_test_performance.png".format(game_name) + "}\n\\centering"
    print(a + b + c)
```
