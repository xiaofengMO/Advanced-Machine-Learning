# get learning curves

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.python.ops import rnn, rnn_cell
import random
import time
import os
import numpy as np
import matplotlib.pyplot as plt
from pylab import savefig

def get_learning_curve(unit_name, mode):
    train_error_list = np.load("../models/RNN_{}_{}_train_error_list.npy".format(unit_name, mode))
    test_error_list = np.load("../models/RNN_{}_{}_test_error_list.npy".format(unit_name, mode))
    train_loss_list = np.load("../models/RNN_{}_{}_train_loss_list.npy".format(unit_name, mode))
    test_loss_list = np.load("../models/RNN_{}_{}_test_loss_list.npy".format(unit_name, mode))
    return train_error_list, test_error_list, train_loss_list, test_loss_list

for unit_name in ["GRU", "LSTM"]:
    for mode in ["32", "64", "128", "stack_32"]:
        try:
            train_error_list, test_error_list, train_loss_list, test_loss_list = get_learning_curve(unit_name, mode)
            plt.figure(num=None, figsize=(4, 3), dpi=128, facecolor='w', edgecolor='k')
            plt.plot(train_error_list)
            plt.plot(test_error_list)
            plt.title("{}, {} error curve".format(unit_name, mode))
            plt.ylabel("error rate")
            plt.xlabel("evaluatin throughout training")
            plt.legend(["training error", "testing error"])
            plt.show()
#             savefig("./image/{}_{}_error_curve.png".format(unit_name, mode))
            
            plt.figure(num=None, figsize=(4, 3), dpi=128, facecolor='w', edgecolor='k')
            plt.plot(train_loss_list)
            plt.plot(test_loss_list)
            plt.title("{}, {} loss curve".format(unit_name, mode))
            plt.ylabel("loss")
            plt.xlabel("evaluatin throughout training")
            plt.legend(["training loss", "testing loss"])
            plt.show()
#             savefig("./image/{}_{}_loss_curve.png".format(unit_name, mode))

        except:
            pass
```
